```python
# This cell creates a function that takes a data frame, and a date-tie column. 
# Then it splits it to three different Series, and return them. 
# The splited columns are: date column, time column, and mMeridiem column. 

date_list, time_list, meridiem_list = [], [], []
def date_time_split(df, column):
    for item in df[column]:
        date, time, meridiem_status = item.split(sep=' ')
        date_list.append(date)
        time_list.append(time)
        meridiem_list.append(meridiem_status)
    date_column = pd.Series(date_list)
    time_column = pd.Series(time_list)
    meridiem_column = pd.Series(meridiem_list)
    return date_column, time_column, meridiem_column
```

```python
# Splitting the 'lpep_pickup_datetime' column to three seperate columns. 
# And after that we drop the 'lpep_pickup_datetime' from the dataset. 

df_final['pickup_date'] = date_time_split(df_final, 'lpep_pickup_datetime')[0]
df_final['pickup_time'] = date_time_split(df_final, 'lpep_pickup_datetime')[1]
df_final['pickup_meridiem'] = date_time_split(df_final, 'lpep_pickup_datetime')[2]
df_final.drop('lpep_pickup_datetime', axis=1, inplace=True)

df_final.head()
```

```python
# Same thing happens for column 'lpep_dropoff_datetime'. 

df_final['dropoff_date'] = date_time_split(df_final, 'lpep_dropoff_datetime')[0]
df_final['dropoff_time'] = date_time_split(df_final, 'lpep_dropoff_datetime')[1]
df_final['dropoff_meridiem'] = date_time_split(df_final, 'lpep_dropoff_datetime')[2]
df_final.drop('lpep_dropoff_datetime', axis=1, inplace=True)

df_final.head()
```

```python
df_final['time_part_of_day'] = pd.Series(np.zeros(len(df_final['pickup_time'])))
```

```python
# Changing the time in to the different part of the day as in below:
# 12:00:00 am - 1:59:59 am      Midnight 
# 2:00:00 am - 5:59:59 am       Early morning
# 6:00:00 am - 8:59:59 am       Morning
# 9:00:00 am - 11:59:59 am      Late morning
# 12:00:00 pm - 3:59:59 pm      Afternoon
# 4:00:00 pm - 4:59:59 pm       Late afternoon
# 5:00:00 pm - 6:59:59 pm       Early evening
# 7:00:00 pm - 8:59:59 pm       Evening 
# 9:00:00 pm - 10:59:59 pm      Late evening
# 11:00:00 pm - 11:59:59 pm     Night


for i in range(len(df_final['pickup_time'])):
    if df_final.pickup_meridiem[i] == 'AM':
        if df_final.pickup_time[i] < '01:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'midnight'
        if '02:00:00' < df_final.pickup_time[i] < '05:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'early morning'
        if '06:00:00' < df_final.pickup_time[i] < '08:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'Morning'
        if '09:00:00' < df_final.pickup_time[i] < '11:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'Late morning'
        if '12:00:00' < df_final.pickup_time[i]:
            df_final.loc[i:i+1,'time_part_of_day'] = 'midnight'
    
    else:
        if df_final.pickup_time[i] < '03:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'afternoon'
        if '04:00:00' < df_final.pickup_time[i] < '04:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'late afternoon'
        if '05:00:00' < df_final.pickup_time[i] < '06:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'early evening'
        if '07:00:00' < df_final.pickup_time[i] < '08:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'evening'
        if '09:00:00' < df_final.pickup_time[i] < '10:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'late evening'
        if '11:00:00' < df_final.pickup_time[i] < '11:59:59':
            df_final.loc[i:i+1,'time_part_of_day'] = 'night'
        if '12:00:00' < df_final.pickup_time[i]:
            df_final.loc[i:i+1,'time_part_of_day'] = 'afternoon'

```
